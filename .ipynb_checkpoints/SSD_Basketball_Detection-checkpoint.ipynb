# SSD Basketball & Player Detection

This notebook trains an SSD (Single Shot Detector) model using torchvision's SSD300 with a VGG16 backbone to detect **BasketBall** and **Player** classes. After training, we evaluate on the test set and generate comparative visualizations.

## 1. Setup & Imports

```python
import os
import cv2
import json
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from pathlib import Path
from collections import defaultdict

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision
from torchvision.models.detection import ssd300_vgg16, SSD300_VGG16_Weights
from torchvision import transforms as T
from torchvision.ops import box_iou

# Device
device = torch.device("mps" if torch.backends.mps.is_available() else "cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Paths
BASE_DIR = Path("archive")
TRAIN_IMG = BASE_DIR / "train" / "images"
TRAIN_LBL = BASE_DIR / "train" / "labels"
VAL_IMG = BASE_DIR / "valid" / "images"
VAL_LBL = BASE_DIR / "valid" / "labels"
TEST_IMG = BASE_DIR / "test" / "images"
TEST_LBL = BASE_DIR / "test" / "labels"

# 3 classes: background (0), BasketBall (1), Player (2)
CLASS_NAMES = ["__background__", "BasketBall", "Player"]
NUM_CLASSES = 3
COLORS = {"BasketBall": "#FF6B35", "Player": "#004E89"}

print(f"Train: {len(list(TRAIN_IMG.glob('*.jpg')))} images")
print(f"Val:   {len(list(VAL_IMG.glob('*.jpg')))} images")
print(f"Test:  {len(list(TEST_IMG.glob('*.jpg')))} images")
```

## 2. Dataset Class

Custom PyTorch dataset to load images and YOLO-format labels, converting them to the format expected by torchvision's SSD.

```python
class BasketballDataset(Dataset):
    """Dataset for basketball/player detection. Converts YOLO format to torchvision format."""
    
    def __init__(self, img_dir, lbl_dir, img_size=300):
        self.img_dir = Path(img_dir)
        self.lbl_dir = Path(lbl_dir)
        self.img_size = img_size
        self.img_files = sorted(list(self.img_dir.glob("*.jpg")))
        
    def __len__(self):
        return len(self.img_files)
    
    def __getitem__(self, idx):
        img_path = self.img_files[idx]
        lbl_path = self.lbl_dir / (img_path.stem + ".txt")
        
        # Load and resize image
        img = cv2.imread(str(img_path))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        orig_h, orig_w = img.shape[:2]
        img = cv2.resize(img, (self.img_size, self.img_size))
        
        # Normalize to [0, 1] and convert to tensor
        img_tensor = torch.from_numpy(img).permute(2, 0, 1).float() / 255.0
        
        # Parse labels
        boxes = []
        labels = []
        if lbl_path.exists():
            with open(lbl_path) as f:
                for line in f:
                    parts = line.strip().split()
                    if len(parts) < 5:
                        continue
                    cls_id = int(parts[0])
                    cx, cy, w, h = float(parts[1]), float(parts[2]), float(parts[3]), float(parts[4])
                    
                    # Convert YOLO normalized to absolute pixel coords at img_size
                    x1 = (cx - w / 2) * self.img_size
                    y1 = (cy - h / 2) * self.img_size
                    x2 = (cx + w / 2) * self.img_size
                    y2 = (cy + h / 2) * self.img_size
                    
                    # Clamp
                    x1 = max(0, x1)
                    y1 = max(0, y1)
                    x2 = min(self.img_size, x2)
                    y2 = min(self.img_size, y2)
                    
                    if x2 > x1 and y2 > y1:
                        boxes.append([x1, y1, x2, y2])
                        labels.append(cls_id + 1)  # +1 because 0 is background in SSD
        
        if len(boxes) == 0:
            boxes = torch.zeros((0, 4), dtype=torch.float32)
            labels = torch.zeros((0,), dtype=torch.int64)
        else:
            boxes = torch.tensor(boxes, dtype=torch.float32)
            labels = torch.tensor(labels, dtype=torch.int64)
        
        target = {"boxes": boxes, "labels": labels}
        return img_tensor, target

def collate_fn(batch):
    images = [item[0] for item in batch]
    targets = [item[1] for item in batch]
    images = torch.stack(images)
    return images, targets

# Create datasets
train_dataset = BasketballDataset(TRAIN_IMG, TRAIN_LBL, img_size=300)
val_dataset = BasketballDataset(VAL_IMG, VAL_LBL, img_size=300)
test_dataset = BasketballDataset(TEST_IMG, TEST_LBL, img_size=300)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, 
                          collate_fn=collate_fn, num_workers=0)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False,
                        collate_fn=collate_fn, num_workers=0)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False,
                         collate_fn=collate_fn, num_workers=0)

print(f"Train batches: {len(train_loader)}")
print(f"Val batches:   {len(val_loader)}")
print(f"Test batches:  {len(test_loader)}")

# Quick sanity check
img, target = train_dataset[0]
print(f"\nSample - Image shape: {img.shape}, Boxes: {target['boxes'].shape}, Labels: {target['labels']}")
```

## 3. Visualize Training Samples

```python
random.seed(42)
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle("SSD Training Samples (300x300 resized)", fontsize=16, fontweight="bold")

indices = random.sample(range(len(train_dataset)), 6)
for ax, idx in zip(axes.flat, indices):
    img, target = train_dataset[idx]
    img_np = img.permute(1, 2, 0).numpy()
    
    ax.imshow(img_np)
    for box, lbl in zip(target["boxes"], target["labels"]):
        x1, y1, x2, y2 = box.numpy()
        cls_name = CLASS_NAMES[lbl.item()]
        color = COLORS[cls_name]
        rect = patches.Rectangle((x1, y1), x2-x1, y2-y1,
                                 linewidth=2, edgecolor=color, facecolor="none")
        ax.add_patch(rect)
        ax.text(x1, y1-3, cls_name, color=color, fontsize=9, fontweight="bold",
                bbox=dict(boxstyle="round,pad=0.2", facecolor="white", alpha=0.7))
    ax.axis("off")

plt.tight_layout()
plt.show()
```

## 4. Build SSD300 Model

We use torchvision's pretrained SSD300 with VGG16 backbone and replace the classification head for our 3 classes (background + BasketBall + Player).

```python
def build_ssd_model(num_classes):
    """Build SSD300 with pretrained VGG16 backbone, replacing the head for our classes."""
    # Load pretrained SSD300
    model = ssd300_vgg16(weights=SSD300_VGG16_Weights.DEFAULT)
    
    # Replace classification head for our number of classes
    # SSD head has classification and regression sub-networks
    in_channels = [c.in_channels for c in model.head.classification_head.module_list]
    num_anchors = [c.out_channels // 91 for c in model.head.classification_head.module_list]  # 91 is COCO classes
    
    # Build new classification head
    new_cls_heads = nn.ModuleList()
    for in_ch, n_anc in zip(in_channels, num_anchors):
        new_cls_heads.append(nn.Conv2d(in_ch, n_anc * num_classes, kernel_size=3, padding=1))
    
    model.head.classification_head.module_list = new_cls_heads
    model.head.classification_head.num_columns = num_classes
    
    return model

model = build_ssd_model(NUM_CLASSES)
model = model.to(device)
print(f"SSD300 model loaded with {NUM_CLASSES} classes")
print(f"Parameters: {sum(p.numel() for p in model.parameters()):,}")
```

## 5. Training Loop

```python
NUM_EPOCHS = 30
LR = 0.001
WEIGHT_DECAY = 5e-4

optimizer = optim.SGD(model.parameters(), lr=LR, momentum=0.9, weight_decay=WEIGHT_DECAY)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.1)

# Training history
history = {
    "train_loss": [], "val_loss": [],
    "train_cls_loss": [], "train_reg_loss": [],
    "val_cls_loss": [], "val_reg_loss": []
}

def train_one_epoch(model, loader, optimizer, device):
    model.train()
    total_loss = 0
    total_cls = 0
    total_reg = 0
    n_batches = 0
    
    for images, targets in loader:
        # SSD expects list of images and list of target dicts
        image_list = [img.to(device) for img in images]
        target_list = [{k: v.to(device) for k, v in t.items()} for t in targets]
        
        # Filter out images with no boxes (SSD requires at least 1 box)
        valid = [(img, tgt) for img, tgt in zip(image_list, target_list) if tgt["boxes"].shape[0] > 0]
        if not valid:
            continue
        image_list = [v[0] for v in valid]
        target_list = [v[1] for v in valid]
        
        losses = model(image_list, target_list)
        loss = sum(losses.values())
        
        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)
        optimizer.step()
        
        total_loss += loss.item()
        total_cls += losses.get("classification", torch.tensor(0)).item()
        total_reg += losses.get("bbox_regression", torch.tensor(0)).item()
        n_batches += 1
    
    return total_loss / max(n_batches, 1), total_cls / max(n_batches, 1), total_reg / max(n_batches, 1)

@torch.no_grad()
def validate(model, loader, device):
    model.train()  # SSD only returns losses in train mode
    total_loss = 0
    total_cls = 0
    total_reg = 0
    n_batches = 0
    
    for images, targets in loader:
        image_list = [img.to(device) for img in images]
        target_list = [{k: v.to(device) for k, v in t.items()} for t in targets]
        
        valid = [(img, tgt) for img, tgt in zip(image_list, target_list) if tgt["boxes"].shape[0] > 0]
        if not valid:
            continue
        image_list = [v[0] for v in valid]
        target_list = [v[1] for v in valid]
        
        losses = model(image_list, target_list)
        loss = sum(losses.values())
        
        total_loss += loss.item()
        total_cls += losses.get("classification", torch.tensor(0)).item()
        total_reg += losses.get("bbox_regression", torch.tensor(0)).item()
        n_batches += 1
    
    return total_loss / max(n_batches, 1), total_cls / max(n_batches, 1), total_reg / max(n_batches, 1)

# Train
best_val_loss = float("inf")
patience_counter = 0
PATIENCE = 7

for epoch in range(NUM_EPOCHS):
    train_loss, train_cls, train_reg = train_one_epoch(model, train_loader, optimizer, device)
    val_loss, val_cls, val_reg = validate(model, val_loader, device)
    scheduler.step()
    
    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)
    history["train_cls_loss"].append(train_cls)
    history["train_reg_loss"].append(train_reg)
    history["val_cls_loss"].append(val_cls)
    history["val_reg_loss"].append(val_reg)
    
    print(f"Epoch {epoch+1:3d}/{NUM_EPOCHS} | "
          f"Train Loss: {train_loss:.4f} (cls: {train_cls:.4f}, reg: {train_reg:.4f}) | "
          f"Val Loss: {val_loss:.4f} (cls: {val_cls:.4f}, reg: {val_reg:.4f})")
    
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), "ssd_best.pth")
        patience_counter = 0
        print(f"  -> Saved best model (val_loss={val_loss:.4f})")
    else:
        patience_counter += 1
        if patience_counter >= PATIENCE:
            print(f"  -> Early stopping at epoch {epoch+1}")
            break

print("\nTraining complete!")
```

## 6. Training Curves

```python
epochs = range(1, len(history["train_loss"]) + 1)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("SSD300 Training Curves", fontsize=16, fontweight="bold")

# Total loss
axes[0].plot(epochs, history["train_loss"], label="Train", color="#FF6B35", linewidth=2)
axes[0].plot(epochs, history["val_loss"], label="Val", color="#004E89", linewidth=2)
axes[0].set_title("Total Loss")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Loss")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Classification loss
axes[1].plot(epochs, history["train_cls_loss"], label="Train", color="#FF6B35", linewidth=2)
axes[1].plot(epochs, history["val_cls_loss"], label="Val", color="#004E89", linewidth=2)
axes[1].set_title("Classification Loss")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("Loss")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# Regression loss
axes[2].plot(epochs, history["train_reg_loss"], label="Train", color="#FF6B35", linewidth=2)
axes[2].plot(epochs, history["val_reg_loss"], label="Val", color="#004E89", linewidth=2)
axes[2].set_title("Regression (Box) Loss")
axes[2].set_xlabel("Epoch")
axes[2].set_ylabel("Loss")
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 7. Evaluate on Test Set

Compute mAP, precision, and recall on the test set.

```python
# Load best model
model.load_state_dict(torch.load("ssd_best.pth", map_location=device, weights_only=True))
model.eval()

def compute_ap(recalls, precisions):
    """Compute Average Precision using 11-point interpolation."""
    ap = 0
    for t in np.arange(0, 1.1, 0.1):
        precs_at_recall = [p for r, p in zip(recalls, precisions) if r >= t]
        ap += max(precs_at_recall) if precs_at_recall else 0
    return ap / 11

def evaluate_ssd(model, dataset, device, conf_thresh=0.3, iou_thresh=0.5):
    """Evaluate SSD model and compute per-class precision, recall, AP."""
    model.eval()
    
    # Collect all predictions and ground truths
    all_preds = defaultdict(list)  # class -> list of (conf, is_tp)
    all_n_gt = defaultdict(int)    # class -> total gt count
    
    all_image_results = []  # for success/failure analysis
    
    for idx in range(len(dataset)):
        img, target = dataset[idx]
        img_input = img.unsqueeze(0).to(device)
        
        with torch.no_grad():
            preds = model(img_input)
        
        pred_boxes = preds[0]["boxes"].cpu()
        pred_labels = preds[0]["labels"].cpu()
        pred_scores = preds[0]["scores"].cpu()
        
        # Filter by confidence
        mask = pred_scores >= conf_thresh
        pred_boxes = pred_boxes[mask]
        pred_labels = pred_labels[mask]
        pred_scores = pred_scores[mask]
        
        gt_boxes = target["boxes"]
        gt_labels = target["labels"]
        
        # Count GTs per class
        for lbl in gt_labels:
            all_n_gt[lbl.item()] += 1
        
        # Match predictions to GT
        matched_gt = set()
        n_matched = 0
        
        for p_idx in range(len(pred_boxes)):
            p_box = pred_boxes[p_idx]
            p_lbl = pred_labels[p_idx].item()
            p_conf = pred_scores[p_idx].item()
            
            best_iou = 0
            best_gt_idx = -1
            for g_idx in range(len(gt_boxes)):
                if gt_labels[g_idx].item() != p_lbl:
                    continue
                if g_idx in matched_gt:
                    continue
                iou = box_iou(p_box.unsqueeze(0), gt_boxes[g_idx].unsqueeze(0))[0, 0].item()
                if iou > best_iou:
                    best_iou = iou
                    best_gt_idx = g_idx
            
            if best_iou >= iou_thresh:
                all_preds[p_lbl].append((p_conf, True))
                matched_gt.add(best_gt_idx)
                n_matched += 1
            else:
                all_preds[p_lbl].append((p_conf, False))
        
        match_rate = n_matched / len(gt_boxes) if len(gt_boxes) > 0 else 0
        avg_conf = pred_scores.mean().item() if len(pred_scores) > 0 else 0
        all_image_results.append({
            "idx": idx, "match_rate": match_rate, "avg_conf": avg_conf,
            "n_gt": len(gt_boxes), "n_pred": len(pred_boxes)
        })
    
    # Compute per-class metrics
    results = {}
    for cls_id in [1, 2]:  # BasketBall, Player
        preds_list = all_preds.get(cls_id, [])
        n_gt = all_n_gt.get(cls_id, 0)
        
        if not preds_list or n_gt == 0:
            results[cls_id] = {"precision": 0, "recall": 0, "ap50": 0, "n_gt": n_gt}
            continue
        
        # Sort by confidence descending
        preds_list.sort(key=lambda x: x[0], reverse=True)
        
        tp_cum = 0
        fp_cum = 0
        precisions = []
        recalls = []
        
        for conf, is_tp in preds_list:
            if is_tp:
                tp_cum += 1
            else:
                fp_cum += 1
            precisions.append(tp_cum / (tp_cum + fp_cum))
            recalls.append(tp_cum / n_gt)
        
        ap = compute_ap(recalls, precisions)
        results[cls_id] = {
            "precision": precisions[-1] if precisions else 0,
            "recall": recalls[-1] if recalls else 0,
            "ap50": ap,
            "n_gt": n_gt
        }
    
    return results, all_image_results

test_results, image_results = evaluate_ssd(model, test_dataset, device)

print("=== SSD300 Test Set Results ===\n")
all_ap = []
for cls_id in [1, 2]:
    cls_name = CLASS_NAMES[cls_id]
    r = test_results[cls_id]
    print(f"{cls_name}:")
    print(f"  Precision: {r['precision']:.4f}")
    print(f"  Recall:    {r['recall']:.4f}")
    print(f"  AP@0.5:    {r['ap50']:.4f}")
    print(f"  GT count:  {r['n_gt']}")
    all_ap.append(r["ap50"])
    print()

mean_ap = np.mean(all_ap)
mean_prec = np.mean([test_results[c]["precision"] for c in [1, 2]])
mean_rec = np.mean([test_results[c]["recall"] for c in [1, 2]])
print(f"Mean AP@0.5:    {mean_ap:.4f}")
print(f"Mean Precision: {mean_prec:.4f}")
print(f"Mean Recall:    {mean_rec:.4f}")
```

## 9. Confidence Distribution

```python
# Collect confidence scores per class
all_confs_ssd = {1: [], 2: []}

model.eval()
for idx in range(len(test_dataset)):
    img, _ = test_dataset[idx]
    with torch.no_grad():
        preds = model([img.to(device)])
    
    for lbl, score in zip(preds[0]["labels"].cpu(), preds[0]["scores"].cpu()):
        if score.item() >= 0.25:
            all_confs_ssd[lbl.item()].append(score.item())

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("SSD300 Confidence Score Distribution on Test Set", fontsize=14, fontweight="bold")

for ax, cls_id in zip(axes, [1, 2]):
    cls_name = CLASS_NAMES[cls_id]
    confs = all_confs_ssd[cls_id]
    ax.hist(confs, bins=30, color=COLORS[cls_name], edgecolor="black", alpha=0.8)
    ax.set_title(f"{cls_name} ({len(confs)} detections)", fontsize=12)
    ax.set_xlabel("Confidence Score")
    ax.set_ylabel("Count")
    if confs:
        ax.axvline(np.mean(confs), color="red", linestyle="--",
                   label=f"Mean: {np.mean(confs):.3f}")
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 10. Success Cases

```python
def show_ssd_predictions(model, dataset, image_indices, device, title="SSD Predictions"):
    """Show SSD predictions vs ground truth."""
    model.eval()
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle(title, fontsize=16, fontweight="bold")
    
    for ax, res in zip(axes.flat, image_indices):
        idx = res["idx"]
        img, target = dataset[idx]
        img_np = img.permute(1, 2, 0).numpy()
        
        ax.imshow(img_np)
        
        # Ground truth (dashed)
        for box, lbl in zip(target["boxes"], target["labels"]):
            x1, y1, x2, y2 = box.numpy()
            cls_name = CLASS_NAMES[lbl.item()]
            color = COLORS[cls_name]
            rect = patches.Rectangle((x1, y1), x2-x1, y2-y1,
                                     linewidth=2, edgecolor=color, facecolor="none", linestyle="--")
            ax.add_patch(rect)
        
        # Predictions (solid)
        with torch.no_grad():
            preds = model([img.to(device)])
        
        for box, lbl, score in zip(preds[0]["boxes"].cpu(), preds[0]["labels"].cpu(), preds[0]["scores"].cpu()):
            if score.item() < 0.3:
                continue
            x1, y1, x2, y2 = box.numpy()
            cls_name = CLASS_NAMES[lbl.item()]
            color = COLORS[cls_name]
            rect = patches.Rectangle((x1, y1), x2-x1, y2-y1,
                                     linewidth=2.5, edgecolor=color, facecolor="none")
            ax.add_patch(rect)
            ax.text(x1, y1-3, f"{cls_name} {score:.2f}", color="white",
                    fontsize=9, fontweight="bold",
                    bbox=dict(boxstyle="round,pad=0.2", facecolor=color, alpha=0.8))
        
        ax.set_title(f"Match: {res['match_rate']:.0%} | GT:{res['n_gt']} Pred:{res['n_pred']}", fontsize=10)
        ax.axis("off")
    
    from matplotlib.lines import Line2D
    legend_elements = [Line2D([0],[0], color="gray", linestyle="--", linewidth=2, label="Ground Truth"),
                       Line2D([0],[0], color="gray", linestyle="-", linewidth=2.5, label="Prediction")]
    fig.legend(handles=legend_elements, loc="lower center", ncol=2, fontsize=12)
    plt.tight_layout(rect=[0, 0.03, 1, 0.97])
    plt.show()

# Sort by match rate for success/failure
image_results.sort(key=lambda x: (x["match_rate"], x["avg_conf"]), reverse=True)
successes = image_results[:6]
failures = image_results[-6:]

show_ssd_predictions(model, test_dataset, successes, device, title="SSD300 SUCCESS Cases (Best Detections)")
```

## 11. Failure Cases

```python
show_ssd_predictions(model, test_dataset, failures, device, title="SSD300 FAILURE Cases (Worst Detections)")
```

## 12. Run on a Single Image with Bounding Boxes

```python
# Single image inference demo
random.seed(123)
sample_idx = random.randint(0, len(test_dataset) - 1)
img, target = test_dataset[sample_idx]

model.eval()
with torch.no_grad():
    preds = model([img.to(device)])

img_np = img.permute(1, 2, 0).numpy()

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(img_np)

n_detections = 0
for box, lbl, score in zip(preds[0]["boxes"].cpu(), preds[0]["labels"].cpu(), preds[0]["scores"].cpu()):
    if score.item() < 0.3:
        continue
    n_detections += 1
    x1, y1, x2, y2 = box.numpy()
    cls_name = CLASS_NAMES[lbl.item()]
    color = COLORS[cls_name]
    
    rect = patches.Rectangle((x1, y1), x2-x1, y2-y1,
                             linewidth=3, edgecolor=color, facecolor="none")
    ax.add_patch(rect)
    ax.text(x1, y1-5, f"{cls_name} {score:.2f}", color="white",
            fontsize=12, fontweight="bold",
            bbox=dict(boxstyle="round,pad=0.3", facecolor=color, alpha=0.9))

ax.set_title(f"SSD300 Detection — Test Image #{sample_idx}", fontsize=14, fontweight="bold")
ax.axis("off")
plt.tight_layout()
plt.show()

print(f"\nDetected {n_detections} objects")
```
